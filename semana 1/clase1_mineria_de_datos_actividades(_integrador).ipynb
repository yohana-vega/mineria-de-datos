### Ejercicio 2: Creación de gráficos en Python

Se trabaja con el siguiente conjunto de datos de ventas:

| Producto | Precio | Cantidad_Vendida |
|----------|--------|------------------|
| A        | 10     | 50               |
| B        | 20     | 30               |
| C        | 15     | 40               |
| D        | 25     | 20               |


Instrucciones:

- Escribir código en Python para crear un histograma de precios.  
- Crear un gráfico de dispersión de Precio vs Cantidad_Vendida.  
- Interpretar los resultados de los gráficos.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear DataFrame
datos = {
    "Producto": ["A", "B", "C", "D"],
    "Precio": [10, 20, 15, 25],
    "Cantidad_Vendida": [50, 30, 40, 20]
}

df = pd.DataFrame(datos)

# Histograma de precios
plt.hist(df["Precio"])
plt.title("Histograma de precios")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()

# Gráfico de dispersión
plt.scatter(df["Precio"], df["Cantidad_Vendida"])
plt.title("Precio vs Cantidad Vendida")
plt.xlabel("Precio")
plt.ylabel("Cantidad Vendida")
plt.show()
```

### Interpretación de los gráficos

El histograma muestra la distribución de los precios de los productos.

El gráfico de dispersión muestra la relación entre el precio y la cantidad vendida. Se observa que a mayor precio, menor cantidad vendida.

Esto permite entender el comportamiento de los datos.




### Ejercicio Integrador: Análisis de clientes (Telecomunicaciones)

1. KDD y CRISP-DM  
KDD: selección de datos, limpieza, transformación, análisis y evaluación.  
CRISP-DM: comprensión del problema, comprensión de los datos, preparación, modelado, evaluación y aplicación.

2. Preprocesamiento  
Revisar datos, detectar errores y valores faltantes.  
Convertir variables categóricas como plan y estado de cuenta.

3. Análisis  
Clientes con mayor consumo tienen planes Premium.  
Clientes con más reclamos tienden a ser morosos.

4. Conclusión  
Identificar clientes con riesgo de morosidad.  
Mejorar el servicio para reducir reclamos.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Datos
datos = {
    "Cliente": [1, 2, 3, 4, 5],
    "Edad": [25, 40, 32, 22, 35],
    "Plan": ["Básico", "Premium", "Estándar", "Básico", "Premium"],
    "Consumo_GB": [5, 50, 10, 7, 45],
    "Reclamos": [1, 0, 2, 3, 0],
    "Estado_Cuenta": ["Pagado", "Pagado", "Moroso", "Moroso", "Pagado"]
}

df = pd.DataFrame(datos)

# Histograma de consumo
sns.histplot(df["Consumo_GB"], bins=5, kde=True)
plt.show()

# Gráfico de barras de estado de cuenta
sns.countplot(x=df["Estado_Cuenta"])
plt.show()
```

##Conclusión
Los datos muestran que existe una relación entre el consumo de datos, los reclamos y el estado de cuenta. Los clientes con menor consumo y más reclamos presentan mayor probabilidad de ser morosos. Esto permite identificar patrones útiles para la toma de decisiones.
