## Análisis exploratorio de Edad y Salario en clientes

### Descripción:
En este ejercicio se realiza un análisis exploratorio de datos utilizando un conjunto de clientes. Se analizan las variables Edad y Salario mediante gráficos univariados y bivariados para identificar la distribución de los datos y posibles relaciones entre ambas variables.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear DataFrame
df = pd.DataFrame({
    "Cliente":       [1, 2, 3, 4, 5],
    "Edad":          [25, 40, 35, 50, 30],
    "Salario":       [50000, 60000, 70000, 80000, 45000],
    "Estado_Cuenta": ["Pagado", "Moroso", "Pagado", "Moroso", "Pagado"]
})
df
```

```python
df = pd.DataFrame({
    "Edad": [25, 30, 45, 35, 50],
    "Salario": [50000, 60000, 70000, 80000, 45000]
})

print("Extracción de datos de las variables a analizar")

df
```

```python
plt.hist(df["Edad"], bins=5, edgecolor="black")
plt.title("Distribución de Edad")
plt.xlabel("Edad")
plt.ylabel("Cantidad de clientes")
plt.yticks(range(0, 3))  # valores enteros visibles
plt.show()

print("El gráfico muestra la distribución de edades agrupadas en intervalos. El eje vertical representa la cantidad de clientes por intervalo.")
```

```python
plt.scatter(df["Edad"], df["Salario"])
plt.title("Edad vs Salario")
plt.xlabel("Edad")
plt.ylabel("Salario")
plt.show()

print("El gráfico muestra la relación entre la edad y el salario. No se observa una relación lineal clara, ya que los salarios no aumentan de forma constante con la edad. Esto indica que no hay una correlación evidente entre ambas variables en este conjunto de datos.")
```

## Conclusión:

### ¿La variable Edad tiene una distribución normal o sesgada?
La variable Edad no presenta una distribución normal. Los datos están dispersos y no muestran una forma simétrica clara, debido a la baja cantidad de observaciones.

### ¿Se observa alguna correlación entre Edad y Salario?
No se observa una correlación clara entre Edad y Salario. Los valores no siguen una tendencia definida, por lo que no se puede afirmar una relación directa entre ambas variables.
