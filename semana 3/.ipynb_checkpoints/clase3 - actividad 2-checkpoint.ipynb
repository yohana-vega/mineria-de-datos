## Actividad 2: Análisis exploratorio de datos con empleados

### Descripción
En esta notebook se realiza un **análisis exploratorio de datos (EDA)** sobre un pequeño conjunto de empleados. Primero se trabaja con un **análisis univariado** de la variable **Salario**, calculando media, mediana, histograma y boxplot. Luego se realiza un **análisis multivariado** entre las variables **Edad** y **Salario**, calculando la correlación y observando la relación mediante un gráfico de dispersión.

El objetivo es interpretar la distribución de los salarios, detectar posibles valores atípicos y analizar si existe relación entre la edad de los empleados y su salario.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 1. Creación del DataFrame
Se crea un conjunto de datos con información básica de empleados: identificador, edad, salario y años de experiencia.


```python
df = pd.DataFrame({
    "Empleado": [1, 2, 3, 4, 5],
    "Edad": [25, 40, 35, 50, 30],
    "Salario": [50000, 60000, 70000, 80000, 45000],
    "Años de Experiencia": [2, 10, 7, 15, 3]
})

df
```

### 2. Análisis univariado de la variable Salario
En esta sección se analiza solamente la variable **Salario**, calculando la media, la mediana y observando su distribución mediante un histograma y un boxplot.


```python
media_salario = df["Salario"].mean()
mediana_salario = df["Salario"].median()

print("Media del salario:", media_salario)
print("Mediana del salario:", mediana_salario)
```

```python
plt.figure(figsize=(7, 4))
sns.histplot(df["Salario"], bins=5, kde=True)
plt.title("Distribución de Salarios")
plt.xlabel("Salario")
plt.ylabel("Frecuencia")
plt.show()
```

**Interpretación del histograma:**  
Los salarios se ubican entre 45.000 y 80.000. Como la cantidad de datos es reducida, no se puede afirmar una distribución normal. Sin embargo, la media y la mediana son muy cercanas, por lo que no se observa un sesgo fuerte en la variable Salario.


```python
plt.figure(figsize=(7, 4))
sns.boxplot(x=df["Salario"])
plt.title("Boxplot de Salarios")
plt.xlabel("Salario")
plt.show()
```

**Interpretación del boxplot:**  
El boxplot permite revisar si hay valores atípicos. En este caso, no se observan puntos separados del resto de la distribución, por lo que no hay outliers importantes en los salarios.


### 3. Análisis multivariado: Edad y Salario
Ahora se analiza la relación entre dos variables: **Edad** y **Salario**. Para eso se calcula la correlación y se visualiza la relación con un gráfico de dispersión.


```python
correlacion = df[["Edad", "Salario"]].corr()
print("Correlación entre Edad y Salario:", round(correlacion.loc["Edad", "Salario"], 3))
correlacion
```

```python
plt.figure(figsize=(7, 4))
sns.scatterplot(data=df, x="Edad", y="Salario")
plt.title("Relación entre Edad y Salario")
plt.xlabel("Edad")
plt.ylabel("Salario")
plt.show()
```

**Interpretación del gráfico de dispersión:**  
El gráfico muestra una tendencia ascendente: a medida que aumenta la edad, también tiende a aumentar el salario. Esto coincide con el valor de correlación obtenido, que indica una relación positiva alta entre ambas variables.


### 4. Preguntas de reflexión y respuestas

**¿La distribución del salario es normal o está sesgada?**  
No se puede afirmar que la distribución sea normal porque el conjunto de datos tiene pocos registros. De todos modos, según el histograma, los salarios se distribuyen entre 45.000 y 80.000 sin una concentración extrema en un solo valor. Además, la media del salario es 61.000 y la mediana es 60.000, valores muy cercanos entre sí. Por eso, en esta muestra no se observa un sesgo fuerte.

**¿Existen valores atípicos en la variable Salario?**  
No se observan valores atípicos importantes. El boxplot no muestra puntos aislados por fuera del rango general de los datos, por lo que los salarios parecen mantenerse dentro de valores esperables para esta muestra.

**¿Qué tipo de correlación hay entre Edad y Salario?**  
La correlación entre Edad y Salario es positiva alta, con un valor aproximado de **0,853**. Esto significa que, dentro de este conjunto de datos, cuando aumenta la edad de los empleados también tiende a aumentar el salario. No implica causalidad, pero sí muestra una relación clara entre ambas variables.

### Conclusión
En esta actividad se realizó un análisis exploratorio de datos sobre empleados. Primero se estudió la variable Salario de forma univariada mediante media, mediana, histograma y boxplot. Luego se analizó la relación entre Edad y Salario mediante correlación y un gráfico de dispersión. A partir del análisis, se observa que los salarios no presentan valores atípicos importantes, no muestran un sesgo fuerte y tienen una relación positiva alta con la edad.
