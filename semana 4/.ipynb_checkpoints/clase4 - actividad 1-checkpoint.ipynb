# Ejercicio 1 – Selección de gráficos

Este ejercicio trata sobre elegir el tipo de gráfico más adecuado según el análisis que se quiere realizar. Se trabajan cuatro casos: comparación de salarios, relación entre variables, detección de valores atípicos y distribución de notas.

## 1. Comparar el salario promedio en diferentes industrias

**Gráfico elegido:** Gráfico de barras.

**Justificación:** El gráfico de barras es adecuado porque permite comparar valores promedio entre categorías. En este caso, cada industria representa una categoría y la altura de cada barra muestra el salario promedio.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_salarios = pd.DataFrame({
    "Industria": ["Tecnología", "Salud", "Educación", "Comercio", "Construcción"],
    "Salario_Promedio": [95000, 80000, 60000, 55000, 70000]
})

sns.barplot(data=df_salarios, x="Industria", y="Salario_Promedio")
plt.title("Salario promedio por industria")
plt.xlabel("Industria")
plt.ylabel("Salario promedio")
plt.xticks(rotation=30)
plt.show()
```

## 2. Analizar relación entre horas trabajadas y productividad

**Gráfico elegido:** Scatter plot o gráfico de dispersión.

**Justificación:** El scatter plot permite observar si existe una relación entre dos variables numéricas. En este caso, ayuda a ver si al aumentar las horas trabajadas también aumenta, disminuye o se mantiene la productividad.

```python
df_productividad = pd.DataFrame({
    "Horas_Trabajadas": [4, 5, 6, 7, 8, 9, 10],
    "Productividad": [50, 58, 65, 70, 72, 71, 68]
})

sns.scatterplot(data=df_productividad, x="Horas_Trabajadas", y="Productividad")
plt.title("Relación entre horas trabajadas y productividad")
plt.xlabel("Horas trabajadas")
plt.ylabel("Productividad")
plt.show()
```

## 3. Detectar valores atípicos en montos de transacciones bancarias

**Gráfico elegido:** Boxplot.

**Justificación:** El boxplot es útil para detectar valores atípicos porque muestra la mediana, los cuartiles y posibles datos extremos. En transacciones bancarias permite identificar montos inusuales que podrían requerir revisión.

```python
df_transacciones = pd.DataFrame({
    "Monto": [1500, 1800, 2000, 2200, 2500, 3000, 3200, 3500, 20000]
})

sns.boxplot(data=df_transacciones, x="Monto")
plt.title("Boxplot de montos de transacciones bancarias")
plt.xlabel("Monto")
plt.show()
```

## 4. Observar la distribución de notas de un curso

**Gráfico elegido:** Histograma.

**Justificación:** El histograma permite observar cómo se distribuyen los valores de una variable numérica. En este caso, ayuda a ver si las notas se concentran en valores bajos, medios o altos.

```python
df_notas = pd.DataFrame({
    "Notas": [4, 5, 6, 6, 7, 7, 7, 8, 8, 9, 10, 6, 5, 7, 8]
})

sns.histplot(data=df_notas, x="Notas", bins=7, kde=True)
plt.title("Distribución de notas de un curso")
plt.xlabel("Notas")
plt.ylabel("Frecuencia")
plt.show()
```

## Conclusión

La selección del gráfico depende del tipo de variables y del objetivo del análisis. Para comparar valores entre categorías se utilizó el gráfico de barras; para analizar relaciones entre variables numéricas, el gráfico de dispersión; para identificar posibles valores atípicos, el boxplot; y para visualizar la distribución de datos, el histograma. Cada tipo de gráfico permite interpretar la información de manera clara según el caso analizado.
