# Ejercicio 3: Gráficos avanzados con Seaborn

En este ejercicio se trabaja con el dataset `diamonds` de Seaborn para aplicar visualizaciones avanzadas. Se analizará la relación entre precio, quilates y profundidad mediante un **Pairplot**, un **Heatmap de correlaciones** y un **Violin Plot** para comparar la distribución del precio según la categoría de corte del diamante.

## Importación de librerías

Se importan las librerías necesarias para trabajar con datos y generar visualizaciones.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

## Carga del dataset

Se carga el dataset `diamonds`, que contiene información sobre diamantes, incluyendo precio, quilates, profundidad, corte y otras variables.

```python
df = sns.load_dataset("diamonds")
df.head()
```

## 1. Pairplot entre precio, quilates y profundidad

Este gráfico permite observar relaciones entre varias variables numéricas al mismo tiempo.

```python
sns.pairplot(df[["price", "carat", "depth"]])
plt.show()
```

## 2. Heatmap de correlaciones

El mapa de calor permite identificar la fuerza de relación entre las variables numéricas del dataset.

```python
correlaciones = df.corr(numeric_only=True)

plt.figure(figsize=(10, 6))
sns.heatmap(correlaciones, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de Correlación de Diamantes")
plt.show()
```

## 3. Violin Plot del precio según el tipo de corte

Este gráfico permite comparar cómo se distribuye el precio de los diamantes en cada categoría de corte.

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x="cut", y="price", data=df, hue="cut", palette="muted", legend=False)
plt.title("Distribución del Precio por Tipo de Corte")
plt.xlabel("Tipo de Corte")
plt.ylabel("Precio")
plt.show()
```

## Conclusión basada en los gráficos

A partir de los gráficos se puede observar que la variable **carat** tiene una relación positiva fuerte con el **precio**: a mayor cantidad de quilates, generalmente mayor es el precio del diamante.

El **heatmap** permite confirmar que el precio está más relacionado con variables como `carat`, `x`, `y` y `z`, que representan características de tamaño del diamante.

El **violin plot** muestra que la distribución del precio varía según el tipo de corte. Algunas categorías presentan mayor concentración de precios bajos, mientras que otras muestran valores más dispersos y precios más altos.
