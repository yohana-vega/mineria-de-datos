# Ejercicio 2: Matplotlib y Seaborn

En este ejercicio se practican dos bibliotecas de visualización en Python: **Matplotlib** y **Seaborn**. Se crea un gráfico de barras para comparar salarios promedio por categoría de empleado y un gráfico de dispersión para analizar la relación entre edad e ingresos.

## 1. Importación de librerías

Se importan las librerías necesarias para trabajar con datos y generar gráficos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Gráfico de barras con Matplotlib

El objetivo es mostrar el salario promedio de diferentes categorías de empleados.

```python
df_salarios = pd.DataFrame({
    "Categoría": ["Junior", "Semi-Senior", "Senior"],
    "Salario_Promedio": [35000, 60000, 90000]
})

df_salarios
```

```python
plt.figure(figsize=(8, 5))
plt.bar(df_salarios["Categoría"], df_salarios["Salario_Promedio"], color=["blue", "green", "red"])
plt.xlabel("Categoría de Empleado")
plt.ylabel("Salario Promedio")
plt.title("Salario Promedio por Categoría")
plt.show()
```

### Interpretación del gráfico de barras

El gráfico permite comparar de forma clara los salarios promedio entre las categorías. Se observa que la categoría **Senior** tiene el salario promedio más alto, seguida por **Semi-Senior** y luego **Junior**.

## 3. Scatterplot con Seaborn

El objetivo es analizar si existe una relación entre la edad y los ingresos de los empleados.

```python
df_empleados = pd.DataFrame({
    "Edad": [25, 32, 40, 35, 50, 45, 29, 37, 41, 60],
    "Ingresos": [30000, 45000, 60000, 55000, 80000, 75000, 40000, 58000, 72000, 90000]
})

df_empleados
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_empleados, x="Edad", y="Ingresos")
plt.title("Relación entre Edad e Ingresos")
plt.xlabel("Edad")
plt.ylabel("Ingresos")
plt.show()
```

### Interpretación del scatterplot

El gráfico muestra una relación positiva entre **Edad** e **Ingresos**. En general, a mayor edad, los ingresos tienden a ser más altos. Esto puede indicar que la experiencia o antigüedad laboral influye en el nivel de ingresos.
