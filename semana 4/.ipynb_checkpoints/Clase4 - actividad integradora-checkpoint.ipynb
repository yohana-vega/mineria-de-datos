## Ejercicio Integrador: Visualización de Datos

En esta notebook se analiza un dataset de ventas de una tienda en línea. El objetivo es aplicar técnicas de visualización de datos vistas en clase para comprender el comportamiento de los clientes, detectar patrones y obtener conclusiones útiles para la toma de decisiones.

## Importación de librerías

Se importan las librerías necesarias para crear el dataset, manipular los datos y generar gráficos.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
```

## Creación del dataset

Se crea un DataFrame con datos de clientes, edad, ingresos, cantidad de compras y categoría.

```python
df = pd.DataFrame({
    "Cliente": [1, 2, 3, 4, 5, 6, 7, 8],
    "Edad": [25, 40, 35, 50, 30, 45, 38, 29],
    "Ingresos": [50000, 60000, 70000, 80000, 45000, 85000, 72000, 48000],
    "Compras": [5, 15, 10, 20, 7, 18, 12, 6],
    "Categoría": ["A", "B", "C", "A", "B", "C", "A", "B"]
})

df
```

## 1. Análisis Exploratorio de Datos (EDA)

En esta sección se generan gráficos básicos para analizar la distribución de las variables y posibles relaciones entre ellas.

### 1.1 Histograma de Edad

Este gráfico permite observar cómo se distribuyen las edades de los clientes.

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["Edad"], bins=5, kde=True)
plt.title("Distribución de Edades")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()
```

### 1.2 Histograma de Ingresos

Este gráfico permite analizar cómo se distribuyen los ingresos de los clientes.

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["Ingresos"], bins=5, kde=True)
plt.title("Distribución de Ingresos")
plt.xlabel("Ingresos")
plt.ylabel("Frecuencia")
plt.show()
```

### 1.3 Boxplot de Compras

El boxplot permite visualizar la dispersión de las compras y detectar posibles valores atípicos.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Compras"])
plt.title("Boxplot de Compras")
plt.xlabel("Cantidad de Compras")
plt.show()
```

### 1.4 Scatter Plot entre Ingresos y Compras

Este gráfico permite analizar si existe una relación entre los ingresos de los clientes y la cantidad de compras realizadas.

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Ingresos", y="Compras", hue="Categoría", data=df, s=100)
plt.title("Relación entre Ingresos y Compras")
plt.xlabel("Ingresos")
plt.ylabel("Compras")
plt.show()
```

## 2. Visualización Avanzada

Se aplican gráficos avanzados para analizar relaciones entre múltiples variables y comparar distribuciones por categoría.

### 2.1 Pairplot

El pairplot permite observar relaciones entre Edad, Ingresos y Compras en un solo conjunto de gráficos.

```python
sns.pairplot(df[["Edad", "Ingresos", "Compras"]])
plt.show()
```

### 2.2 Heatmap de correlaciones

El mapa de calor muestra la relación entre las variables numéricas del dataset.

```python
correlaciones = df.corr(numeric_only=True)

plt.figure(figsize=(8, 5))
sns.heatmap(correlaciones, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de Correlación de Variables")
plt.show()
```

### 2.3 Violin Plot de Compras por Categoría

El violin plot permite comparar cómo se distribuye la cantidad de compras según la categoría del cliente.

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x="Categoría", y="Compras", data=df, hue="Categoría", palette="muted", legend=False)
plt.title("Distribución de Compras por Categoría")
plt.xlabel("Categoría")
plt.ylabel("Compras")
plt.show()
```
